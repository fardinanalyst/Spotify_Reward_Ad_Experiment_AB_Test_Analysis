```python
def keep_columns(df: pd.DataFrame, columns_to_keep: list) -> pd.DataFrame:
    existing_cols = [col for col in columns_to_keep if col in df.columns]
    missing = set(columns_to_keep) - set(existing_cols)
    if missing:
        print(f"⚠️ Warning: These columns were not found and will be skipped: {missing}")
    return df[existing_cols].copy()
```

```python
import pandas as pd

# 1️⃣ Aggregate conversions
daily_conversion_rate = (
    baseline_df[baseline_df['conversion_date'].notna()]
    .groupby('conversion_date')['user_id']
    .nunique()
    .reset_index(name='converted_users')
)

# 2️⃣ Aggregate free users per day
daily_free_users = (
    baseline_df[baseline_df['is_premium'] == 0]
    .groupby('session_date')['user_id']
    .nunique()
    .reset_index(name='free_users')
)

# 3️⃣ Ensure both are datetime
daily_conversion_rate['conversion_date'] = pd.to_datetime(daily_conversion_rate['conversion_date']).dt.floor('D')

# Convert session_date to just date
daily_free_users['session_date'] = pd.to_datetime(daily_free_users['session_date']).dt.floor('D')

# 4️⃣ Merge on date
daily_conversion_rate = daily_conversion_rate.merge(
    daily_free_users,
    left_on='conversion_date',
    right_on='session_date',
    how='left'
)

# 5️⃣ Drop duplicate date column
daily_conversion_rate.drop(columns=['session_date'], inplace=True)

# 6️⃣ Calculate daily conversion rate
daily_conversion_rate['conversion_rate'] = (
    daily_conversion_rate['converted_users'] / daily_conversion_rate['free_users']
)

print(daily_conversion_rate)
```

```python

x1 = 609      # conversions in control
n1 = 10000    # total users in control

x2 = 738      # conversions in treatment
n2 = 10000    # total users in treatment

p1 = x1 / n1
p2 = x2 / n2

print("Control conversion rate:", p1)
print("Treatment conversion rate:", p2)
```

```python
p_pool = (x1 + x2) / (n1 + n2)
print("Pooled proportion:", p_pool)
```

```python
import math

SE = math.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))
print("Standard Error:", SE)
```

```python
Z = (p1 - p2) / SE
print("Z-statistic:", Z)
```

```python
from scipy.stats import norm

p_value = 2 * (1 - norm.cdf(abs(Z)))
print("p-value:", p_value)
```

```python
# Effect Size: Cramér’s V
n = contingency.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(contingency.shape) - 1)))
print("Cramér's V Effect Size:", cramers_v)
```

```python
# ✅ Ensure session_date exists
AB_df['session_date'] = pd.to_datetime(AB_df['session_start']).dt.date

# ✅ Define signup date as first session date
AB_cohort_df = AB_df.groupby('user_id', as_index=False).agg(
    signup_date=('session_date', 'min'),
    experiment_group=('experiment_group', 'first')
)

# ✅ All session activity
AB_activity_df = AB_df[['user_id', 'session_date']].drop_duplicates()
AB_activity_df = AB_activity_df.merge(AB_cohort_df, on='user_id', how='left')

# ✅ Days since signup
AB_activity_df['session_date'] = pd.to_datetime(AB_activity_df['session_date'])
AB_activity_df['signup_date'] = pd.to_datetime(AB_activity_df['signup_date'])
AB_activity_df['days_since_signup'] = (AB_activity_df['session_date'] - AB_activity_df['signup_date']).dt.days

# ✅ Retention by experiment group (7 and 14-day)
retention = AB_activity_df[AB_activity_df['days_since_signup'].isin([7, 14])]
retention_summary = (
    retention.groupby(['experiment_group', 'days_since_signup'])['user_id']
    .nunique()
    .reset_index(name='retained_users')
)

# ✅ Merge with cohort sizes
cohort_sizes = AB_cohort_df.groupby('experiment_group')['user_id'].nunique().reset_index(name='total_users')
retention_summary = retention_summary.merge(cohort_sizes, on='experiment_group')
retention_summary['retention_rate'] = retention_summary['retained_users'] / retention_summary['total_users']

print(retention_summary)
```

```python
import pandas as pd
from scipy.stats import ttest_ind
import numpy as np

# Step 1 — Session-level dataset (avg ads seen in each session per user)
session_df = (
    AB_df.groupby(['user_id', 'session_id'], as_index=False)
         .agg(num_ads_per_session=('num_ads', 'sum'))
)

# Step 2 — Attach experiment group (unique per user)
session_df = session_df.merge(
    AB_df[['user_id', 'experiment_group']].drop_duplicates(),
    on='user_id',
    how='left'
)

# Split into groups
control_ads = session_df[session_df['experiment_group']=="control"]['num_ads_per_session']
treatment_ads = session_df[session_df['experiment_group']=="treatment"]['num_ads_per_session']

# Welch’s t-test
t_stat, p_value = ttest_ind(control_ads, treatment_ads, equal_var=False)

# Effect Size: Cohen's d
cohen_d = (
    (control_ads.mean() - treatment_ads.mean()) /
    np.sqrt((control_ads.std()**2 + treatment_ads.std()**2) / 2)
)

print("Control Avg Ads/Session:", control_ads.mean())
print("Treatment Avg Ads/Session:", treatment_ads.mean())
print("Welch t-statistic:", t_stat)
print("p-value:", p_value)
print("Cohen's d:", abs(cohen_d))

```
